```python
import pandas as pd
```

```python
fintech = pd.read_excel("http://glebvasiliev.com/fintech.xlsx", sheet_name="Coinbase")
fintech
```

```python
fintech2 = pd.read_excel("http://glebvasiliev.com/fintech.xlsx", sheet_name="PayPal")
fintech2
```

```python
fintech.assign(ma10 = fintech["Adj Close"].rolling(10).mean())

fintech.set_index("Date").assign(ma2w = lambda df: df["Adj Close"].rolling("14D").mean())

fintech.assign(running_mean = fintech["Adj Close"].expanding().mean())
fintech.assign(cumulative_sum = fintech["Adj Close"].cumsum())

(
fintech
    .set_index("Date")
    .assign(diff = lambda df: df["Adj Close"].diff(),
           diff2 = lambda df: df["Adj Close"].diff(2),
           pct_change = lambda df: df["Adj Close"].pct_change(),
           prev_adj_close = lambda df: df["Adj Close"].shift(1),
           prev2_adj_close = lambda df: df["Adj Close"].shift(2),
           next_adj_close = lambda df: df["Adj Close"].shift(-1))    
)
```

```python
pd.merge_ordered(fintech, fintech2, on="Date")
pd.merge_ordered(fintech, fintech2, on="Date", fill_method="ffill")
#.ffill()

pd.merge_asof(fintech, fintech2, on="Date")
pd.merge_asof(fintech2, fintech, on="Date")
pd.merge_asof(fintech2, fintech, on="Date", direction="forward")
```
