```python
climate = pd.read_excel("http://glebvasiliev.com/climate.xls")
```

```python
climate
climate.describe()
climate.info()
```

```python
climate
climate.melt(id_vars=climate.columns[0:6])
climate.melt(id_vars=climate.columns[0:6], var_name="Year", value_name="Indicator")
```

```python
(
climate
    .melt(id_vars=climate.columns[0:6], var_name="Year", value_name="Indicator")
    .dropna()
    #.drop()
    .pivot_table(index=["Country code", "Country name", "SCALE", "Decimals", "Year"],
                columns=["Series code", "Series name"],
                values="Indicator",
                aggfunc="first")
    .reset_index()
    .droplevel(level=0, axis=1)
)
```

```python
spotify = pd.read_csv("http://glebvasiliev.com/spotify23.csv", encoding="latin-1")
```

```python
#spotify.info()
#Series.astype()
spotify = spotify.assign(streams = pd.to_numeric(spotify["streams"], errors="coerce"))
```

```python
spotify
pd.crosstab(spotify["key"], spotify["mode"])
pd.crosstab(spotify["key"], spotify["mode"], normalize=True)
pd.crosstab(spotify["key"], spotify["mode"], normalize="columns")
pd.crosstab(spotify["key"], spotify["mode"], normalize="index")
pd.crosstab(spotify["key"], spotify["mode"], values=spotify["streams"], aggfunc="mean")
```
