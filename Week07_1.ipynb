```python
import pandas as pd
import numpy as np
```

```python
airports = pd.read_csv("http://glebvasiliev.com/airports.csv")
airlines = pd.read_csv("http://glebvasiliev.com/airlines.csv")
routes = pd.read_csv("http://glebvasiliev.com/routes.csv")
fleet = pd.read_csv("http://glebvasiliev.com/fleet.csv")
```

```python
airports.info()
airports
```

```python
(
routes
    .merge(airlines
           .assign(AirlineID = airlines["Airline ID"].astype("str")),
           left_on="airline ID",
           right_on="AirlineID")
)

(
routes
    .merge(airlines
           .assign(AirlineID = airlines["Airline ID"].astype("str")),
           left_on="airline ID",
           right_on="AirlineID",
           how="left")
)


(
routes
    .rename(columns=lambda x: x.strip())
    .merge(airlines
           .assign(AirlineID = airlines["Airline ID"].astype("str")),
           left_on="airline ID",
           right_on="AirlineID",
           how="left")
    .merge(airports, left_on="source airport", right_on="IATA")
)
```
