```python
import pandas as pd
import numpy as np
```

```python
spotify = pd.read_csv("http://glebvasiliev.com/spotify23.csv", encoding="latin-1")
```

```python
spotify.danceability.mean()
spotify.danceability.median()
spotify.loc[:, ["danceability", "bpm"]].mean()
spotify.danceability.std()
spotify.danceability.quantile(0.25)
spotify.artist_names.mode()
```

```python
spotify.artist_names.unique()

spotify2 = spotify.query("artist_names in ['Dua Lipa', 'OneRepublic', 'Miley Cyrus']")
```

```python
spotify2.groupby("artist_names")
spotify2.groupby("artist_names").get_group("Miley Cyrus")

spotify3 = spotify2.groupby("artist_names")
spotify3.describe()
#spotify2.describe()
spotify3.head(1)
spotify3.tail(1)
spotify3.nth(2)
spotify3.groups
spotify3.size()
spotify3.nunique()
```

```python
spotify2.groupby("artist_names").mean()
spotify2.groupby("artist_names")[["danceability", "acousticness"]].mean()
spotify2.groupby("artist_names")[["danceability", "acousticness"]].agg(np.mean)
spotify2.groupby("artist_names", as_index=False)[["danceability", "acousticness"]].agg(np.mean)
```

```python
(
spotify
    .groupby("artist_names")
    [["danceability", "energy"]]
    .agg(["mean", "count", "first"])
    .droplevel(0, axis=1)
    .set_axis(["danceability", "count_d", "first_d", "energy", "count_e", "first_e"], axis=1)
    .query("count_d > 1")
    .loc[:, ["danceability", "energy"]]
    .sort_values(["danceability", "energy"], ascending=[False, False])
    .query("danceability > 70 & energy > 70")
)
```

```python
#dataframe.quantile(0.75) - dataframe.quantile(0.25)

(
spotify
    .groupby("artist_names")
    [["danceability", "energy"]]
    .agg(lambda x: x.quantile(0.75) - x.quantile(0.25))
)


#( this is still to be fixed!
#spotify
#    .groupby("artist_names")
#    .filter(lambda g: len(g) > 1)
#    [["danceability", "energy"]]
#    .agg("mean")
#    .sort_values(["danceability", "energy"], ascending=[False, False])
#    .query("danceability > 70 & energy > 70")
#)
```
