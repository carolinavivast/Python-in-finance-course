```python
sp500["signalOne"] = np.where(sp500["previousClose"] < sp500["twoHundredDayAverage"],
                              "buy", "-")
sp500
sp500.sort_values(["industry", "payoutRatio"], ascending=[True, False])
#sp500 = sp500.sort_values(["industry", "payoutRatio"], ascending=[True, False])
#sp500.sort_values(["industry", "payoutRatio"], ascending=[True, False], inplace=True)
sp500

#incorrect: sp500.signalttrtr = np.where(sp500["previousClose"] < sp500["twoHundredDayAverage"],"buy", "-")

sp500 = sp500.assign(signalOne = np.where(sp500["previousClose"] < sp500["twoHundredDayAverage"],
                              "buy", "-"),
            signalTwo = np.where(sp500["previousClose"] < sp500["fiftyDayAverage"],
                              "buy", "-"))
```

```python
sp500.query("signalOne == 'buy' & signalTwo == 'buy'")
sp500.query("signalOne == 'buy' | signalTwo == 'buy'")
sp500.query("ticker in ['ABBV', 'MMM']")
sp500.query("ticker not in ['ABBV', 'MMM']")
sp500.query("ticker == ['ABBV', 'MMM']")
sp500.query("index > 100")
```

```python
sp500.drop("signalOne", axis=1)
sp500
sp500.drop_duplicates("industry").sort_values("industry")
sp500.drop_duplicates("industry").sort_values("industry").rename(columns={"signalOne":"mainSignal"})
```
